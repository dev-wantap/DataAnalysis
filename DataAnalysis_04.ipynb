# 상자 그래프 (boxplot)

```python
import csv
import matplotlib.pyplot as plt


f = open('Data/SeoulTemp.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
next(data)
#header = ['날짜', '지점', '평균기온(°C)', '최저기 온(°C)', '최고기온(°C)']
jan = []
aug = []



for row in data:
    if (row[0].split("-")[1] == '01'):
        jan.append(float(row[-1]))
    if (row[0].split("-")[1] == '08'):
        aug.append(float(row[-1]))

plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.unicode_minus'] = False
        
plt.boxplot([jan, aug])
plt.xticks([1, 2], ['1월', '8월'])
plt.show()
f.close
```

## 서울의 월별 최고 기온 분석

```python
import csv
import matplotlib.pyplot as plt

f = open('Data/SeoulTemp.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
next(data)
#header = ['날짜', '지점', '평균기온(°C)', '최저기 온(°C)', '최고기온(°C)']
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[0].split('-')[0] != 1978:
        month_index = int(row[0].split('-')[1]) - 1
        month[month_index].append(float(row[-1]))

plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.unicode_minus'] = False

plt.boxplot(month)
plt.title("서울 월별 최고 기온 분석")
plt.figure(dpi = 100)

plt.show()
```

## 서울 일별 최고 기온 분석 (8월)

```python
import csv
import matplotlib.pyplot as plt

f = open('Data/SeoulTemp.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
next(data)
#header = ['날짜', '지점', '평균기온(°C)', '최저기 온(°C)', '최고기온(°C)']
date = []
for i in range(31):
    date.append([])

for row in data:
    if row[0].split('-')[0] != 1978:
        date_index = int(row[0].split('-')[2]) - 1
        date[date_index].append(float(row[-1]))

plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.unicode_minus'] = False

plt.boxplot(date)
plt.title("안동 지역 일별 최고 기온 분석")

plt.show()
```
