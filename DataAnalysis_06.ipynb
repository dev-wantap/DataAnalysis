## 지역별 남녀 인구 구조 파악

```python
import csv
import matplotlib.pyplot as plt

f = open('population_202005.csv', encoding = 'cp949')
data = csv.reader(f)
popu = []
town_name = input("행정구역 (읍/면/동) : ")

for row in data:
    if (town_name in row[0]):
        for i in row[3:]:
            popu.append(int(i))
        break

plt.bar(range(101), popu)
plt.show()
f.close()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('Data/population_202005_gender.csv', encoding = 'cp949')
data = csv.reader(f)
male = []
female = []
town_name = input("행정구역 (읍/면/동) : ")

for row in data:
    if (town_name in row[0]):
        for i in row[3:104]:
            male.append(int(i))
        for i in row[106:]:
            female.append(int(i))
        break

plt.style.use('ggplot')
plt.figure(figsize = (4,8), dpi = 100)
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
plt.barh(np.arange(0, 101, step=1), male, color = 'salmon', label = '남성', alpha = 1)
plt.barh(np.arange(0, 101, step=1), female, color = 'skyblue', label = '여성', alpha = 1)
plt.legend()
plt.title(town_name+"의 인구 분포")
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('Data/population_202005_gender.csv', encoding = 'cp949')
data = csv.reader(f)
male = []
female = []
town_name = input("행정구역 (읍/면/동) : ")

for row in data:
    if (town_name in row[0]):
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:]:
            female.append(int(i))
        break

plt.style.use('ggplot')
plt.figure(figsize = (4,8), dpi = 100)
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
plt.barh(np.arange(0, 101, step=1), male, color = 'salmon', label = '남성', alpha = 1)
plt.barh(np.arange(0, 101, step=1), female, color = 'skyblue', label = '여성', alpha = 1)
plt.legend()
plt.title(town_name+"의 인구 분포")
plt.show()

f.close()
```

### 모든 운영체제에서 한글이 보이게 하는 설정

```python
import platform
from matplotlib import font_manager, rc 
import matplotlib

# 운영 체제마다 한글이 보이게 하는 설정

# 윈도우
if platform.system() == 'Windows':
    path = "c:\Windows\Fonts\malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name() 
    rc('font', family=font_name)

#맥
elif platform.system() == 'Darwin': 
    rc('font', family='AppleGothic')

# 리눅스
elif platform.system() == 'Linux':
    rc('font', family='NanumBarunGothic')
```
