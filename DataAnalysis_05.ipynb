```python
import csv
import matplotlib.pyplot as plt

t1 = 0
t2 = 0

f = open('population_202005.csv', encoding = 'cp949')
data = csv.reader(f)
popu = []
town1_name = input("행정구역1 (읍/면/동) : ")
town2_name = input("행정구역2 (읍/면/동) : ")
town1 = []
town2 = []

for row in data:
    if (town1_name in row[0] and t1 != 1):
        for i in row[3:]:
            town1.append(int(i))
        t1 = 1
    if (town2_name in row[0] and t2 != 1):
        for i in row[3:]:
            town2.append(int(i))
        t2 = 1
    if (t1 == 1 and t2 == 1):
        break


f.close()

plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 12

plt.style.use('ggplot')
plt.plot(town1, color = 'skyblue', label = town1_name)
plt.plot(town2, color = 'hotpink', label = town2_name)
plt.legend()
plt.title(town1_name+"|"+town2_name+"의 인구구조")
plt.show()
```
