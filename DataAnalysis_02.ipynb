## 라인 그래프 꾸미기

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 13), [27, 1, 7, 14 ,19, 23, 24, 26, 20, 13, 8, 3], 'hotpink')
plt.show()
```

## 제목 추가

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 12
A = [27, 1, 7, 14 ,19, 23, 24, 26, 20, 13, 8, 3]
B = [5, 10, 17, 7 ,10, 30, 22, 17, 13, 6, 2, 0]
plt.plot(range(1, 13), A, linestyle='--', color='hotpink', marker = 'o', label = 'A지점')
plt.plot(range(1, 13), B, linestyle='-', color='skyblue', marker = 'o', label = 'B지점')

plt.title('민원 발생 추이')
plt.legend()
plt.show()
```

## x축 눈금 지정

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams['font.size'] = 14

A = [27, 1, 7, 14 ,19, 23, 24, 26, 20, 13, 8, 3]
B = [5, 10, 17, 7 ,10, 30, 22, 17, 13, 6, 2, 0]
xAxis = range(1,13)

plt.plot(xAxis, A, linestyle='-.', color='hotpink', marker = 'o', label = 'A지점')
plt.plot(xAxis, B, linestyle=':', color='skyblue', marker = 'o', label = 'B지점')

plt.title('민원 발생 추이')
plt.legend()
plt.xticks(xAxis)
plt.show()
```
